# Data Visualization I — Titanic Dataset

**Objective:**
1. Use Seaborn's built-in **Titanic** dataset (891 rows) and explore patterns in the data through visualizations.
2. Plot a **histogram** to see how ticket fare is distributed across passengers.

### About the dataset
The Titanic dataset records information about the 891 passengers from the RMS Titanic, including age, sex, ticket class, fare, port of embarkation, and whether they survived. It's a classic dataset for learning data analysis and visualization.

## Step 1: Import Libraries

````python
# pandas — DataFrame operations
import pandas as pd

# numpy — numerical operations
import numpy as np

# matplotlib.pyplot — base plotting library; seaborn builds on top of this
import matplotlib.pyplot as plt

# seaborn — statistical visualization library; cleaner, more attractive plots
import seaborn as sns

# Display plots inline in the notebook
%matplotlib inline

# Set seaborn's default style — gives a nice grid background and color palette
sns.set_style('whitegrid')

print("Libraries imported successfully.")
````

## Step 2: Load the Titanic Dataset

````python
# Seaborn ships with several built-in datasets accessible via load_dataset().
# This downloads the Titanic CSV from seaborn's GitHub repo (needs internet on first call).
df = sns.load_dataset('titanic')

# shape -> (rows, columns)
print("Shape:", df.shape)

# head() shows the first 5 rows
df.head()
````

````python
# info() — column names, non-null counts, data types
df.info()
````

````python
# describe() — summary statistics for numeric columns
df.describe()
````

## Step 3: Visualize Patterns in the Data

We'll explore relationships between variables using different Seaborn plots.

### 3.1 Survival Count

````python
# countplot — shows the count of each category
# A simple way to see how many passengers survived vs did not
plt.figure(figsize=(6, 4))
sns.countplot(x='survived', data=df, palette='Set2')
plt.title('Survival Count (0 = Did Not Survive, 1 = Survived)')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.show()
````

### 3.2 Survival by Sex

````python
# Use 'hue' to break the count into sub-categories — here, by sex
# This reveals the famous "women and children first" pattern
plt.figure(figsize=(7, 4))
sns.countplot(x='survived', hue='sex', data=df, palette='Set1')
plt.title('Survival Count by Sex')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.show()
````

### 3.3 Survival by Passenger Class

````python
# pclass = ticket class (1 = first class, 2 = second, 3 = third)
# This shows the influence of socio-economic status on survival
plt.figure(figsize=(7, 4))
sns.countplot(x='pclass', hue='survived', data=df, palette='coolwarm')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.show()
````

### 3.4 Age Distribution

````python
# histplot — shows distribution of a single numeric variable
# kde=True overlays a smooth curve estimating the density
plt.figure(figsize=(8, 4))
sns.histplot(df['age'].dropna(), kde=True, color='skyblue', bins=30)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
````

### 3.5 Age vs Fare (Scatter Plot)

````python
# scatterplot — shows the relationship between two numeric variables
# Color-coded by survival status to look for patterns
plt.figure(figsize=(8, 5))
sns.scatterplot(x='age', y='fare', hue='survived', data=df, palette='Set1', alpha=0.7)
plt.title('Age vs Fare, colored by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
````

### 3.6 Boxplot — Age by Passenger Class

````python
# boxplot — shows the distribution of a numeric variable across categories
# Box = middle 50% (Q1 to Q3), line in box = median, dots = outliers
plt.figure(figsize=(8, 5))
sns.boxplot(x='pclass', y='age', data=df, palette='pastel')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
````

### 3.7 Heatmap — Correlation between numeric features

````python
# Pick only numeric columns and compute pairwise correlations
numeric_df = df.select_dtypes(include=['number'])

# heatmap — displays a 2D table as a colored grid
# annot=True writes the correlation value in each cell
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
````

## Step 4: Histogram of Fare Distribution

The **histogram** is the standard plot for showing how a single continuous variable is distributed. The x-axis is divided into bins (intervals), and each bar shows how many observations fall into that bin.

````python
# Basic histogram of the fare column
plt.figure(figsize=(10, 5))
sns.histplot(df['fare'], bins=30, color='teal', kde=False)
plt.title('Distribution of Ticket Fare')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
````

````python
# Histogram with KDE (Kernel Density Estimate) — smooth curve overlay
# KDE gives a smoothed view of the distribution shape
plt.figure(figsize=(10, 5))
sns.histplot(df['fare'], bins=30, kde=True, color='purple')
plt.title('Fare Distribution with Density Curve')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
````

````python
# Quick statistical summary of fare to interpret the histogram
print("Fare statistics:")
print(df['fare'].describe())
````

**Observations from the fare histogram:**
- The fare distribution is **right-skewed** (long tail to the right) — most passengers paid low fares while a small number paid very high fares.
- The majority of fares fall under £30.
- A few passengers paid extreme amounts (over £200), reflecting first-class luxury suites.
- This skewness matches what we'd expect: most passengers traveled in third class with cheap tickets, and only a few in first class with very expensive ones.

### Bonus: Fare Distribution by Passenger Class

````python
# Use 'hue' to split the histogram by passenger class
# Reveals that 1st class fares are much higher than 2nd or 3rd class
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='fare', hue='pclass', bins=30, kde=True, palette='viridis', multiple='stack')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
````

# Data Visualization I — What We Actually Did

**Big picture:** Up until now, all our notebooks were about *manipulating* data — cleaning it, transforming it, building models on it, evaluating those models. This notebook is fundamentally different: it's about **looking** at the data. Not summarizing with numbers, not predicting with a model — just rendering the data as pictures so a human can spot patterns by eye.

This matters more than people realize. The single most famous example in statistics is **Anscombe's Quartet** — four datasets that have *identical* mean, variance, correlation, and regression line, but look completely different when you plot them. One is linear, one is curved, one has an outlier, one is just garbage. If you only looked at the summary statistics, you'd think they were the same dataset. The moment you plot them, the differences are obvious.

That's the whole point of Exploratory Data Analysis (EDA): **before you build any model, look at your data.** You'll catch outliers, see relationships, notice missing patterns, and form hypotheses that pure number-crunching would miss. The Titanic dataset is the textbook playground for this because it's small enough to plot every variable but rich enough to contain real, interpretable patterns about a real historical event.

The dataset records 891 passengers from the RMS Titanic — their age, sex, ticket class, fare paid, port of embarkation, family members aboard, and most importantly, whether they survived the sinking on April 15, 1912. The famous "women and children first" protocol means survival wasn't random — it correlated heavily with sex, age, and ticket class. The whole goal of this notebook is to **see** those correlations as pictures.

## Why Seaborn?

Python has two main plotting libraries:

- **Matplotlib** — the granddaddy, hugely powerful, but verbose. To make a nice statistical plot you write a lot of code.
- **Seaborn** — built *on top* of Matplotlib. It comes with sensible defaults, attractive color palettes, and high-level functions for statistical plots (`countplot`, `boxplot`, `histplot`, `heatmap`). One line of seaborn often replaces ten lines of matplotlib.

Both work together — you import both because seaborn handles the statistical plot but matplotlib handles things like figure size, titles, labels. The standard pattern is `import matplotlib.pyplot as plt` for general control plus `import seaborn as sns` for the actual plot calls.

### Step 4 — The Fare Histogram (Question 2)

This is what the assignment specifically asks for. A histogram of fare shows three things at once: the **range** of values (min to max), the **shape** of the distribution (symmetric, skewed, bimodal?), and where the **bulk** of values cluster.

We did three versions:
- **Plain histogram** — just `sns.histplot(df['fare'], bins=30)`. Bars show how many passengers fall into each fare bin.
- **With KDE curve** — adds a smooth density estimate on top, useful for visualizing the distribution shape without the jagged edges of binned bars.
- **Split by passenger class** — using `hue='pclass'` and `multiple='stack'` to color each fare bar by class. Reveals that first-class fares span a huge range while third-class is tightly packed at the low end.

**The key observation:** the fare distribution is **strongly right-skewed**. A right-skewed distribution has a long tail extending to the right — most values are low, but a few extreme high values pull the tail out. Specifically:
- The mean fare is around £32, but the median is only about £14
- Most passengers paid under £30
- A small handful paid over £200 — the highest recorded fare is £512.33 (a first-class luxury suite)

This skewness reflects the social reality of 1912: the Titanic was an elite ocean liner, and while it had a third-class section (the bulk of passengers), the famous luxury was concentrated in a small number of very expensive cabins.

### What "bins" mean

A histogram divides the x-axis range into equal-width bins (intervals). Each bar's height = count of values falling in that bin. The number of bins matters:
- Too few bins (e.g., 5) — the histogram looks blocky and hides detail
- Too many bins (e.g., 200) — the histogram looks noisy and jagged
- Sweet spot — usually 20–50 bins for most datasets; we used 30

There are mathematical rules (Sturges' rule, Freedman-Diaconis rule) for picking optimal bin counts, but visually choosing 20–50 works fine for most cases.

# Possible Viva Questions

### Conceptual

**Q1. What is data visualization?**
The graphical representation of data using charts, plots, and graphs to make patterns, trends, and relationships easier to understand than from raw numbers alone.

**Q2. Why do we need data visualization?**
Humans are visual creatures — we spot patterns in pictures faster than in tables. Visualization helps us identify outliers, relationships, distributions, trends, and data quality issues that would be invisible in summary statistics.

**Q3. What is the difference between Matplotlib and Seaborn?**
Matplotlib is the foundational plotting library — powerful but verbose. Seaborn is built on top of Matplotlib and provides high-level functions for statistical plots with attractive defaults. Seaborn is concise; Matplotlib gives full control.

**Q4. What is Exploratory Data Analysis (EDA)?**
The initial process of investigating data using visualizations and summary statistics to discover patterns, spot anomalies, test hypotheses, and check assumptions before formal modeling.

**Q5. Why is Seaborn preferred for statistical visualization?**
Built-in support for statistical plots (boxplot, violinplot, heatmap, regplot), automatic handling of categorical data via `hue`, attractive default color palettes, and integration with pandas DataFrames.

### Plot Types

**Q6. What is a histogram?**
A plot that shows the distribution of a continuous variable by dividing its range into bins and showing the count (or frequency) of values in each bin as bars.

**Q7. What is the difference between a histogram and a bar chart?**
A bar chart shows counts of *categorical* variables (with gaps between bars). A histogram shows the distribution of a *continuous numeric* variable (bars are adjacent because bins are continuous intervals).

**Q8. What is KDE (Kernel Density Estimate)?**
A non-parametric way of estimating the probability density function of a continuous variable. It produces a smooth curve that approximates what the histogram is showing, without the binning artifacts.

**Q9. What is a boxplot? What does it show?**
A plot that summarizes a distribution using five numbers: minimum (whisker bottom), Q1 (box bottom), median (line in box), Q3 (box top), maximum (whisker top). Points beyond the whiskers (1.5×IQR) are shown as dots and considered outliers.

**Q10. What is a scatter plot?**
A plot showing the relationship between two numeric variables, with one variable on each axis and each data point as a dot.

**Q11. What is a heatmap?**
A 2D visualization where values are represented by color intensity. Commonly used to show correlation matrices or pivot tables.

**Q12. What is a countplot?**
A bar chart showing the count (frequency) of each category in a categorical variable. Equivalent to `value_counts()` plotted as bars.

**Q13. What is the purpose of the `hue` parameter in seaborn?**
It splits the plot by a third categorical variable, drawing separate-colored bars/lines/points for each category. Lets you see how a relationship varies across groups.

### Histograms Specifically

**Q14. What is "skewness"?**
A measure of distribution asymmetry. Right-skewed (positive skew) = long tail on the right (most values small, few very large). Left-skewed (negative skew) = long tail on the left. Symmetric = bell-shaped.

**Q15. Was the fare distribution skewed? Why?**
Yes — strongly right-skewed. Most passengers were in third class with cheap tickets; a small number paid premium first-class fares (up to £512). This pulls the tail far to the right.

**Q16. How does the number of bins affect a histogram?**
Too few → the plot loses detail and hides features. Too many → the plot looks noisy and jagged. Typically 20–50 bins work well; rules like Sturges' or Freedman-Diaconis give a starting point.

**Q17. Why did you use `kde=True`?**
To overlay a smooth density curve on top of the histogram bars. The KDE shows the shape of the distribution without the noise of bin boundaries.

**Q18. What does it mean if mean > median in a distribution?**
The distribution is right-skewed — extreme high values are pulling the mean above the median. (Mean < median = left-skewed.)

### Titanic Findings

**Q19. What was the overall survival rate?**
About 38% — roughly 342 of 891 passengers survived.

**Q20. Which gender had a higher survival rate?**
Females — by a large margin. About 74% of women survived versus only ~19% of men. This reflects the "women and children first" lifeboat protocol.

**Q21. How did passenger class affect survival?**
First-class passengers had the highest survival rate (~63%), second class about 47%, third class only about 24%. Class was strongly negatively correlated with survival.

**Q22. Why is `pclass` *negatively* correlated with survival?**
Because lower numbers represent higher classes — class 1 is first class, class 3 is third class. Higher class number = lower social status = lower survival. The negative correlation captures this.

**Q23. What was the typical age of passengers?**
Most passengers were 20–40 years old; the median age was about 28. The age column had ~177 missing values (about 20% of rows).

**Q24. What pattern did you see in the age-by-class boxplot?**
First-class passengers were older (median ~37), second class around 29, third class around 24. Wealthier passengers tended to be more established adults.

### Code-Specific

**Q25. What does `sns.load_dataset('titanic')` do?**
Downloads the Titanic CSV from seaborn's GitHub repo and returns it as a pandas DataFrame. Requires internet on the first call.

**Q26. Why did you write `df['age'].dropna()` before plotting?**
Because the age column has missing values (NaN), and the histogram function can't plot NaN — it would error out or produce a misleading bar. `dropna()` filters them out.

**Q27. Why use `plt.figure(figsize=(...))` before each plot?**
To control the size of the plot. Without it, plots use the default size which is often too small for clear viewing.

**Q28. What does `multiple='stack'` do in `histplot`?**
When using `hue` to split a histogram into groups, this option stacks the bars on top of each other rather than overlaying them. Other options: `'layer'` (overlapping), `'dodge'` (side-by-side), `'fill'` (normalized to 100%).

**Q29. What is a "palette" in seaborn?**
A set of colors used to render categories in a plot. Examples: `'Set1'`, `'Set2'`, `'pastel'`, `'viridis'`, `'coolwarm'`. You can also pass a list of custom colors.

**Q30. What did the correlation heatmap reveal?**
Strong negative correlation between `pclass` and `fare` (higher class = higher fare). Negative correlation between `pclass` and `survived` (higher class = better chance). Positive correlation between `fare` and `survived`. Weak correlations elsewhere.

### Higher-Level

**Q31. What's the difference between a histogram and a KDE plot?**
A histogram is discrete (binned bars); a KDE is continuous (smooth curve). KDE doesn't depend on bin choices, but it does depend on bandwidth choices.

**Q32. What's the difference between a boxplot and a violin plot?**
Both show distribution across categories. A boxplot shows quartiles and outliers as a box+whiskers shape. A violin plot adds a KDE curve on each side of the box, showing the full shape of the distribution — it's a boxplot + density combined.

**Q33. When would you use a heatmap vs a scatter plot?**
Heatmap: when you have a 2D matrix of values to display (correlation matrix, pivot table). Scatter plot: when you have two variables and want to see individual data points and their relationship.

**Q34. What is the rule for outlier detection in a boxplot?**
A point is flagged as an outlier if it's more than 1.5 × IQR (interquartile range) above Q3 or below Q1. IQR = Q3 − Q1.

**Q35. How would you visualize three numeric variables together?**
- 3D scatter plot (matplotlib's `mpl_toolkits.mplot3d`)
- Or a 2D scatter where the third variable is encoded as point size or color (`sns.scatterplot(x=..., y=..., size=..., hue=...)`)
- Or a `pairplot` showing all pairwise scatter plots in a grid.

## Plot 1 — Survival Count (Countplot)

**Code:** `sns.countplot(x='survived', data=df)`

### What this plot is

A **countplot** is essentially a bar chart that shows how many times each category appears in a column. You give it a categorical column (one with discrete values), and it draws one bar per category, with the bar's height equal to the count of rows in that category.

In our case, `survived` has only two values: `0` (did not survive) and `1` (survived). So we get exactly two bars.

### How to read it

- **X-axis:** the categories (0 and 1)
- **Y-axis:** the count of rows in each category
- **Bar height:** number of passengers in that category

### What we observed

- Bar at 0 (didn't survive): tall — about 549 passengers
- Bar at 1 (survived): shorter — about 342 passengers
- Conclusion: ~38% of passengers survived. The disaster killed nearly two-thirds of those onboard.

### When to use a countplot

Use `countplot` whenever you want to **count occurrences of categorical values** — like how many users are in each gender, how many products of each color, how many emails of each priority. It's the visual equivalent of `df['column'].value_counts()`.

### Why not a regular bar chart?

Matplotlib's `bar()` requires you to compute counts yourself first. Seaborn's `countplot` does the counting + plotting in one step. Convenience.

## Plot 2 — Survival by Sex (Grouped Countplot)

**Code:** `sns.countplot(x='survived', hue='sex', data=df)`

### What this plot is

This is the same `countplot` as before, but now we add `hue='sex'`. The `hue` parameter tells seaborn to **subdivide each bar into multiple colored sub-bars**, one per category of the hue variable.

So instead of one bar for "didn't survive" and one for "survived", we now get four bars:
- Didn't survive — male
- Didn't survive — female
- Survived — male
- Survived — female

The bars are placed side-by-side within each survival group, with different colors for males vs females.

### How to read it

- **X-axis:** primary category (survived 0 or 1)
- **Y-axis:** count
- **Color:** secondary category (male vs female)
- **Side-by-side bars:** comparison within each x-axis group

### What we observed

- Among the "didn't survive" group: the **male bar is huge**, the female bar is small
- Among the "survived" group: the **female bar is taller** than the male bar
- Translation: women survived at much higher rates than men. The lifeboat protocol "women and children first" is visible in raw counts.

### Why this is the most informative single plot in the dataset

Survival on the Titanic is essentially a function of three variables: sex, class, and age. Plotting survival broken down by sex reveals the single strongest predictor in the entire dataset. If you had to predict survival from just one variable, it would be sex.

### When to use grouped countplots

Whenever you want to **compare counts across two categorical variables simultaneously**. Examples: customer signups by month and by source, products sold by region and by category, students by year and by major.

## Plot 4 — Age Distribution (Histogram with KDE)

**Code:** `sns.histplot(df['age'].dropna(), kde=True, bins=30)`

### What this plot is

A **histogram** shows the distribution of a single continuous numeric variable. Here's how it works:

1. Take all the age values (range: 0 to 80 years)
2. Divide that range into equal-width bins (we used 30 bins, so each bin spans about 80/30 ≈ 2.7 years)
3. Count how many ages fall into each bin
4. Draw a bar for each bin, height = count

The **KDE (Kernel Density Estimate)** is a smooth curve overlaid on top. It estimates what the underlying continuous distribution would look like if you had infinitely many samples. Think of it as a "smoothed-out histogram."

### How to read it

- **X-axis:** the value of the variable (age in years)
- **Y-axis:** count (or density when KDE is shown)
- **Bar height:** how many passengers fall in that age range
- **KDE curve:** smoothed estimate of the distribution shape

### What we observed

- **Peak around 20–35 years** — this is where most passengers fall
- **Long right tail** extending into 60s and 70s — some elderly passengers
- **Small bump near 0–5** — infants and young children
- The curve is roughly bell-shaped with a slight right skew

### Why we wrote `.dropna()`

The age column has 177 missing values (NaN). Histograms can't plot NaN — they'd error or skip silently. `.dropna()` filters them out so the plot shows only known ages.

### Why use histograms

To answer questions like:
- What's the most common value?
- Is the distribution symmetric or skewed?
- Are there multiple peaks (modes)?
- Are there outliers?
- What's the range of values?

A histogram answers all of these visually in one shot.

### Bins — why 30?

The number of bins is a tradeoff:
- Too few bins (5 or 10) → blocky, hides detail, oversmooths
- Too many bins (200) → noisy, jagged, exaggerates noise
- 20–50 is the sweet spot for most datasets
- 30 is a safe default

There are formal rules (Sturges' rule says use log₂(n)+1 bins; Freedman-Diaconis uses 2·IQR/n^(1/3)), but eyeballing usually works.

## Plot 5 — Age vs Fare Scatter (Colored by Survival)

**Code:** `sns.scatterplot(x='age', y='fare', hue='survived', data=df)`

### What this plot is

A **scatter plot** places one dot for each row in the dataset, with the dot's x-position determined by one variable and y-position by another. The `hue='survived'` parameter colors each dot based on survival status.

### How to read it

- **X-axis:** age
- **Y-axis:** fare
- **Each dot:** one passenger
- **Dot color:** red = didn't survive, blue = survived (or whatever the palette assigns)
- **`alpha=0.7`** makes dots semi-transparent so overlapping dots become visible

### What to look for in a scatter plot

1. **Trend** — do the dots form a line or curve? (Linear or non-linear relationship)
2. **Direction** — when one variable goes up, does the other go up (positive) or down (negative)?
3. **Strength** — are the dots tightly clustered along the trend, or scattered widely?
4. **Outliers** — any dots far from the rest?
5. **Clusters** — do certain dot colors group in specific areas?

### What we observed in this specific plot

- **Most dots are at the bottom** (low fares) — confirming the right-skew of fare
- **A few extreme high-fare outliers** — fares above £200 are very rare
- **No strong age-fare correlation** — the dots don't form a clear line; older passengers didn't necessarily pay more
- **Survivor dots (one color) cluster more in the higher-fare region** — yet another confirmation that fare correlates with survival

### When to use scatter plots

The default plot for **two continuous numeric variables**. Use it whenever you want to see the relationship, correlation, or pattern between two such variables.

### What scatter plots can't do

They can't show how *many* dots are in dense regions. If 200 passengers all have age=25 and fare=8, they all overlap into a single dot. Workarounds: `alpha` for transparency (so overlapping dots darken), `jitter` to add small random noise, or use a 2D histogram (`hexbin` or `kdeplot`) instead.

## Plot 6 — Boxplot of Age by Passenger Class

**Code:** `sns.boxplot(x='pclass', y='age', data=df)`

### What this plot is

A **boxplot** (also called box-and-whisker plot) is one of the densest, most informative summary plots in statistics. In a single shape, it shows:

- **Median** — the line inside the box (50th percentile)
- **Q1 (25th percentile)** — the bottom of the box
- **Q3 (75th percentile)** — the top of the box
- **IQR (Interquartile Range)** — the box height, equal to Q3 − Q1
- **Whiskers** — extend to the smallest and largest values *within 1.5×IQR* of the box
- **Outliers** — points beyond the whiskers, drawn as dots

### How to read it

For each class, you see one boxplot. To compare:
- **Median lines** — which class has older passengers?
- **Box sizes** — which class has more spread?
- **Whisker lengths** — which class has wider extremes?
- **Outlier dots** — are there unusual points?

### What we observed

- **Class 1:** highest median age (~37), broader spread, few outliers
- **Class 2:** median around 29, moderate spread
- **Class 3:** lowest median (~24), tighter spread but possibly more young children at the bottom
- The pattern: wealthier passengers were typically older, established adults; third class included more young workers and emigrant families

### Why boxplots are powerful

They show **five summary statistics at once** for **multiple categories simultaneously** in a single figure. A table of statistics (mean/median/Q1/Q3 for three classes) requires reading 12+ numbers; a boxplot conveys it instantly.

### When to use a boxplot

When you want to **compare a numeric variable across categories** — sales by region, test scores by school, salary by department, etc. It's the gold standard for "how does X differ between groups?"

### What boxplots can't show

The actual **shape** of the distribution. Two distributions with identical quartiles but very different shapes (one bimodal, one unimodal) look identical in a boxplot. **Violin plots** solve this by adding a KDE curve on each side of the box.

## Plot 7 — Correlation Heatmap

**Code:** `sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')`

### What this plot is

A **heatmap** displays a 2D matrix of values as a colored grid, where color encodes magnitude. We computed the **correlation matrix** of all numeric columns — a square table where each cell shows how strongly two variables move together.

**Correlation values range from −1 to +1:**
- **+1** = perfectly positive (when X goes up, Y always goes up by the same proportion)
- **0** = no linear relationship
- **−1** = perfectly negative (when X goes up, Y always goes down)

### How to read it

- **Rows and columns:** the variables (same set on both axes)
- **Diagonal cells:** always 1 (a variable's correlation with itself)
- **Color:** red for positive, blue for negative, white for near-zero (with the `coolwarm` palette)
- **Number annotations:** `annot=True` writes the actual correlation value in each cell

### What we observed

- **`pclass` ↔ `fare`: strong negative** — higher class number (= lower class) means lower fare. Makes sense: third class is cheaper than first.
- **`pclass` ↔ `survived`: negative** — third class died more
- **`fare` ↔ `survived`: positive** — wealthier passengers survived more
- **`age` ↔ `pclass`: negative** — older passengers were in higher classes
- **`sibsp` ↔ `parch`: positive** — passengers traveling with siblings often had parents/children too (families)

### Why heatmaps are the EDA workhorse

Before any modeling, looking at the correlation heatmap tells you:
- Which variables are strongly related (good predictors)
- Which variables are redundant (high correlation = multicollinearity, can hurt regression)
- Which pairs are surprising and worth investigating

### Limitations of correlation

Correlation measures **linear** relationships only. Two variables can be perfectly related in a nonlinear way (parabolic, sinusoidal, etc.) and still show correlation = 0. Always pair correlation analysis with scatter plots.

### When to use a heatmap

Whenever you have a 2D table of numbers to display visually:
- Correlation matrices (most common)
- Confusion matrices (we saw this in Naïve Bayes/Logistic notebooks)
- Pivot tables (e.g., sales by month × region)
- Distance matrices

## Plot 8 — Histogram of Fare (THE main question)

**Code:** `sns.histplot(df['fare'], bins=30)`

### What this plot is

This is the assignment's core question: **how is fare distributed?** Same `histplot` as in Plot 4, but for the fare column instead of age.

### How to read it

- **X-axis:** fare in British pounds
- **Y-axis:** count of passengers
- Each bar = number of passengers whose fare fell in that bin

### What we observed

- **Massive spike at low values** (£0–£20) — most passengers paid little
- **Steady drop-off** as fare increases
- **Long thin tail** extending to £500+ — a few extreme premium fares
- **Almost no passengers at intermediate values** like £100–£200

This is a textbook **right-skewed distribution**. The mean (£32) is much higher than the median (£14) because the few extreme high values pull the mean upward.

### Why fare is so skewed

The Titanic was a luxury ocean liner with a small number of premium suites and a much larger third-class section. The premium suites cost 30–50× more than the cheap tickets. So the *typical* fare was low, but the *maximum* fare was extremely high. Whenever you have this kind of "many small + few enormous" pattern, you get right skew.

### Real-world examples of right-skewed distributions

- **Income** — most people earn moderate amounts, a few earn millions
- **City populations** — most cities are small, a few are huge
- **House prices** — most are middle-class, a few are mansions
- **Word frequencies in text** — most words are rare, a few ("the", "a") are everywhere

### Why this matters for analysis

Right-skewed data violates the normality assumption of many statistical methods (linear regression, t-tests, ANOVA). Common fix: apply a **log transformation** (which we did in Data Wrangling II), which compresses the long tail and pulls the distribution toward symmetry.

## Plot 8b — Fare Histogram with KDE Curve

**Code:** `sns.histplot(df['fare'], kde=True, bins=30)`

### What this adds

A smooth density curve overlaid on top of the histogram bars. The curve estimates what the underlying continuous distribution looks like, ignoring the artifacts of bin boundaries.

### How to read the KDE

The y-axis switches from "count" to "density" (so the area under the curve = 1.0). The shape of the curve reveals:
- Where the peak is (mode)
- Whether there's one peak (unimodal) or multiple (bimodal/multimodal)
- How sharp or gradual the drop-off is

For our fare data, the KDE curve shows a sharp peak near zero with a long, smooth tail extending right — confirming the right-skew visually.

### Why use KDE alongside histograms

- Histograms can be sensitive to bin choice
- KDE gives a "bin-free" view
- Together they show both the discrete reality (bars) and the smooth shape (curve)

## Plot 8c — Fare Distribution by Passenger Class

**Code:** `sns.histplot(data=df, x='fare', hue='pclass', bins=30, multiple='stack')`

### What this plot adds

Splits the fare histogram into three colored sub-distributions, one per passenger class, stacked vertically. So at any fare bin, you see how many class-1, class-2, and class-3 passengers paid in that range.

### How to read it

- **X-axis:** fare
- **Y-axis:** total count, with each bar segmented by class
- **Bar colors:** different shade per class

### What we observed

- **Third class (largest segment):** entirely concentrated at the low end (under £30)
- **Second class:** mostly clustered around £10–£30
- **First class:** spread across a wide range, from ~£25 up to £500+
- The high-fare tail is essentially **all first class** — second/third class never paid those amounts

This single plot encapsulates the whole socioeconomic story of the Titanic: three nearly disjoint fare distributions, with first class being not just more expensive but vastly more *variable* in price.

### `multiple='stack'` explained

When using `hue` with histograms, you can choose how to display overlapping groups:
- **`'stack'`** — bars stacked on top of each other (total height = total count)
- **`'layer'`** — bars overlap, partially transparent
- **`'dodge'`** — bars side-by-side at each bin
- **`'fill'`** — normalized so each bin sums to 100% (shows proportions)

We chose `'stack'` because it makes the total fare distribution shape visible while still showing class composition.

# The Big Picture — Why Each Plot Type Exists

Different plots answer different questions:

| Question | Best plot |
|----------|-----------|
| "How many of each category?" | Countplot |
| "How is this numeric variable distributed?" | Histogram (+ KDE) |
| "Are these two numeric variables related?" | Scatter plot |
| "How does this numeric variable vary across categories?" | Boxplot (or violin) |
| "How do all my variables correlate with each other?" | Heatmap |
| "How does outcome A vary by category B?" | Grouped countplot or stacked bar |

Memorize this table and you'll never be stuck choosing a plot.

# General Tips for Reading Any Plot

When you encounter any visualization, walk through these questions:

1. **What's on the axes?** Always read x and y labels first.
2. **What's the unit?** Counts? Percentages? Density?
3. **What's the scale?** Linear or log? Did the analyst transform anything?
4. **What's the shape?** Symmetric, skewed, bimodal, uniform?
5. **Are there outliers?** Points clearly separated from the rest?
6. **What's the trend?** Increasing, decreasing, flat, cyclic?
7. **What's the spread?** Tight cluster or wide variation?
8. **What's missing?** Could the chart be hiding something (zero baseline, cropped axes, cherry-picked range)?

# Common Pitfalls to Watch For

- **Truncated y-axis:** if a bar chart starts at 50 instead of 0, small differences look huge
- **Misleading colors:** using red for "positive" and green for "negative" can confuse colorblind viewers
- **Cherry-picked ranges:** plotting only 2010–2020 of a 50-year trend
- **Overplotting:** dense scatter plots where dots overlap into a single blob (use alpha or hexbin)
- **Pie charts with too many slices:** humans can't compare angles well — use bars instead
- **3D plots that aren't necessary:** they distort and rarely help
