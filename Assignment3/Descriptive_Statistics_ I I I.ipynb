### Key Concepts

**Measures of Central Tendency** — describe the *center* of a distribution:
- **Mean** — arithmetic average; sensitive to outliers.
- **Median** — middle value when sorted; robust to outliers.
- **Mode** — most frequent value.

**Measures of Variability** — describe the *spread* of a distribution:
- **Minimum / Maximum** — smallest / largest value.
- **Range** — max − min.
- **Standard Deviation (std)** — average distance of each value from the mean.
- **Variance** — square of the standard deviation.
- **Percentile** — value below which a given % of observations fall (25th = Q1, 50th = median, 75th = Q3).

## Step 1: Import Libraries

```python
# pandas — DataFrame operations: read_csv(), groupby(), describe(), agg().
import pandas as pd

# numpy — numerical operations and array support (used internally by pandas).
import numpy as np

# Display all columns in output
pd.set_option('display.max_columns', None)
```

## Step 2: Load the Dataset

```python
# read_csv() loads a CSV file into a pandas DataFrame.
df = pd.read_csv('Descriptive_Statistics_ I I I.csv')
# df1 = pd.read_csv('Descriptive_Statistics_ I I I.csv')  # IGNORE
# shape -> (rows, columns)
print("Shape:", df.shape)

# head() shows the first 5 rows.
df.head()
```

```python
# dtypes shows the data type of each column:
#   float64 -> numeric (continuous)
#   object  -> string (categorical 'species')
print(df.dtypes)
# print(df.dtypes)
# Verify the categorical variable's unique categories.
print("\nSpecies categories:", df['species'].unique())
```

## Part 1: Summary Statistics Grouped by a Categorical Variable

We treat **`species`** as the categorical (qualitative) variable and the four flower measurements as the numeric (quantitative) variables. We then compute summary statistics for each numeric variable *within each species*.

This is exactly analogous to the textbook example "income grouped by age groups" — we just have "petal/sepal measurements grouped by species".

### 1.1 Group by Species and Compute Summary Statistics

```python
# groupby('species') splits the DataFrame into 3 groups (one per species).
# Calling an aggregation function then computes that statistic for each group.

# MEAN — arithmetic average of each numeric column, per species.
print("MEAN of each numeric variable, grouped by species:\n")
print(df.groupby('species').mean(numeric_only=True))
```

```python
# MEDIAN — middle value (50th percentile) of each numeric column, per species.
# More robust than mean when the data has outliers.
print("MEDIAN of each numeric variable, grouped by species:\n")
print(df.groupby('species').median(numeric_only=True))
```

```python
# MIN — smallest value in each group.
print("MINIMUM of each numeric variable, grouped by species:\n")
print(df.groupby('species').min(numeric_only=True))
```

```python
# MAX — largest value in each group.
print("MAXIMUM of each numeric variable, grouped by species:\n")
print(df.groupby('species').max(numeric_only=True))
```

```python
# STANDARD DEVIATION — measures spread of values around the mean within each group.
# A small std means values are tightly clustered; a large std means they are spread out.
print("STANDARD DEVIATION of each numeric variable, grouped by species:\n")
print(df.groupby('species').std(numeric_only=True))
```

### 1.2 All Statistics in One Table — agg()

```python
# agg() applies multiple aggregation functions at once and presents them in
# one tidy table. Far more compact than calling mean(), median(), etc. separately.
summary = df.groupby('species').agg(['mean', 'median', 'min', 'max', 'std'])

print("Combined summary statistics grouped by species:")
summary
```

### 1.3 Create a List of Numeric Values for Each Categorical Response

The assignment asks: *Create a list that contains a numeric value for each response to the categorical variable.*

We build a list where each element is the **mean `sepal_length`** of one species. So the list has exactly 3 numbers — one numeric value per category (species).

```python
# tolist() converts a pandas Series into a regular Python list.
# Here we get the mean sepal_length for each of the 3 species as a list.
mean_sepal_length_per_species = df.groupby('species')['sepal_length'].mean().tolist()

print("Categories (species):", df['species'].unique().tolist())
print("Mean sepal_length per species:", mean_sepal_length_per_species)
```

```python
# We can also build a dictionary {category -> numeric value} for clearer mapping.
# to_dict() turns a Series into a Python dict.
species_means = df.groupby('species')['sepal_length'].mean().to_dict()
print("Mean sepal_length per species (as dict):")
for species, mean_val in species_means.items():
    print(f"  {species}: {round(mean_val, 3)}")
```

## Part 2: Statistical Details for Each Iris Species

The assignment asks for percentile, mean, and standard deviation for each species. The fastest way is `describe()`, which gives all of these (count, mean, std, min, 25%, 50%, 75%, max) in one call.

### 2.1 Iris-setosa

```python
# Filter the DataFrame to only Iris-setosa rows.
# df['species'] == 'Iris-setosa' returns a boolean mask;
# df[mask] selects the rows where the mask is True.
setosa = df[df['species'] == 'Iris-setosa']

# describe() returns: count, mean, std, min, 25%, 50% (median), 75%, max.
# That single call covers ALL the statistics the assignment asks for.
print("Statistical details for Iris-setosa:")
setosa.describe()
```

### 2.2 Iris-versicolor

```python
versicolor = df[df['species'] == 'Iris-versicolor']
print("Statistical details for Iris-versicolor:")
versicolor.describe()
```

### 2.3 Iris-virginica

```python
virginica = df[df['species'] == 'Iris-virginica']
print("Statistical details for Iris-virginica:")
virginica.describe()
```

### 2.4 One-line Alternative — All Three at Once

```python
# Same result for ALL three species at once via groupby + describe.
df.groupby('species').describe()
```
